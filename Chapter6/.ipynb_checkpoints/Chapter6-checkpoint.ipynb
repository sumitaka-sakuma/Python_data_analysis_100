```python
import pandas as pd
```

```python
# 生産工場
factories = pd.read_csv("../Chapter6_csv/tbl_factory.csv", index_col=0)
factories
```

```python
# 倉庫
warehouses = pd.read_csv("../Chapter6_csv/tbl_warehouse.csv", index_col=0)
warehouses
```

```python
# 倉庫と工場間の輸送コスト
cost = pd.read_csv("../Chapter6_csv/rel_cost.csv", index_col=0)
cost.head()
```

```python
# 2019年の向上への部品輸送実績
trans = pd.read_csv("../Chapter6_csv/tbl_transaction.csv", index_col=0)
trans.head()
```

```python
join_data = pd.merge(trans, cost, left_on=["ToFC", "FromWH"], right_on=["FCID", "WHID"], how="left")
join_data.head()
```

```python
# 工場データの付与
join_data = pd.merge(join_data, factories, left_on="ToFC", right_on="FCID", how="left")
join_data.head()
```

```python
# 倉庫情報の付与
join_data = pd.merge(join_data, warehouses, left_on="FromWH", right_on="WHID", how="left")
join_data = join_data[["TransactionDate", "Quantity", "Cost", "ToFC", "FCName", "FCDemand", "FromWH", "WHName", "WHSupply", "WHRegion"]]
join_data.head()
```

```python
# 関東支社のみのデータ
kanto = join_data.loc[join_data["WHRegion"]=="関東"]
kanto.head()
```

```python
# 東北支社のみのデータ
tohoku = join_data.loc[join_data["WHRegion"]=="東北"]
tohoku.head()
```

```python
# 輸送実績の総コスト
print("関東支社の総コスト:" + str(kanto["Cost"].sum()) + "万円")
print("東北支社の総コスト:" + str(tohoku["Cost"].sum()) + "万円")
```

```python
# 輸送実績の総輸送部品個数
print("関東支社の総輸送部品個数:" + str(kanto["Quantity"].sum()) + "個")
print("東北支社の総輸送部品個数:" + str(tohoku["Quantity"].sum()) + "個")
```

```python
# 部品一つ当たりの輸送コスト
tmp = (kanto["Cost"].sum() / kanto["Quantity"].sum()) * 10000
print("関東支社の部品一つ当たりの輸送コスト：" + str(int(tmp)) + "円")

tmp = (tohoku["Cost"].sum() / tohoku["Quantity"].sum()) * 10000
print("東北支社の部品一つ当たりの輸送コスト：" + str(int(tmp)) + "円")
```

```python
# 支社ごとの平均輸送コスト
cost_chk = pd.merge(cost, factories, on="FCID", how="left")

print("東京支社の平均輸送コスト：" + str(cost_chk["Cost"].loc[cost_chk["FCRegion"]=="関東"].mean()) + "万円")
print("東北支社の平均輸送コスト：" + str(cost_chk["Cost"].loc[cost_chk["FCRegion"]=="東北"].mean()) + "万円")
```

```python
# ネットワークの可視化
import networkx as nx
import matplotlib.pyplot as plt

# グラフオブジェクトの作成
G = nx.Graph()

# 頂点の設定
G.add_node("nodeA")
G.add_node("nodeB")
G.add_node("nodeC")

# 辺の設定
G.add_edge("nodeA", "nodeB")
G.add_edge("nodeA", "nodeC")
G.add_edge("nodeB", "nodeC")

# 座標の設定
pos = {}
pos["nodeA"] = (0, 0)
pos["nodeB"] = (1, 1)
pos["nodeC"] = (0, 1)

# 描画
nx.draw(G, pos)

# 表示
plt.show()
```

```python
# ネットワークに頂点を追加する
G.add_node("nodeD")

G.add_edge("nodeA", "nodeD")

pos["nodeD"] = (1, 0)
nx.draw(G, pos, with_labels=True)
```

```python
# ルートの重みづけ
import numpy as np

# データの読み込み
df_w = pd.read_csv("../Chapter6_csv/network_weight.csv")
df_p = pd.read_csv("../Chapter6_csv/network_pos.csv")

# エッジの重みリスト化
size = 10
edge_weights = []
for i in range(len(df_w)):
    for j in range(len(df_w.columns)):
        edge_weights.append(df_w.iloc[i][j]*size)

# グラフオブジェクトの作成
G = nx.Graph()

# 頂点の設定
for i in range(len(df_w.columns)):
    G.add_node(df_w.columns[i])

# 辺の設定
for i in range(len(df_w.columns)):
    for j in range(len(df_w.columns)):
        G.add_edge(df_w.columns[i], df_w.columns[j])

# 座標の設定
pos = {}
for i in range(len(df_w.columns)):
    node = df_w.columns[i]
    pos[node] = (df_p[node][0], df_p[node][1])
    
# 描画
nx.draw(G, pos, with_labels=True, font_size=16, node_size=1000, node_color='k', font_color='w', width=edge_weights)

# 表示
plt.show()
```
